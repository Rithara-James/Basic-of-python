PRIME NUMBERS

```python
def is_prime(num):
  if num <=1:
    return False
  for i in range(2 , num):
    if (num%2 == 0):
      return False
  return True

number = int(input("enter the number"))

if is_prime(number):
  print(number,"is a prime number")
else:
  print(number,"is not a prime number")
```

PRODUCT OF RANDOM NUMBER

```python
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

user_ans = int(input(f"what is {num1}x{num2}="))

if user_ans == num1 * num2:
  print("correct,Well done")
else:
  print("incorrect,Sorry try again")
```

WORD COUNTER

```python
text = "i am red in colour with around shape and i grow on tree and is commonly eat raw and i am who famous by phone brand and i also famous in story of isaac netwon and gravity. "
words = text.split()
word_count = {}
for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1
for word, count in word_count.items():
  print(f"{word}:{count}")
```

CHECK PALINDROME

```python

def is_palindrome(str):
  str = str.lower()
  clean_text = ''.join(char for char in str if char.isalnum() )
  return clean_text == clean_text[::-1]

input_text = "racecar"
print(is_palindrome(input_text))

input_text2 = "A man, a plan, a canal, Panama"
print(is_palindrome(input_text2))
```
